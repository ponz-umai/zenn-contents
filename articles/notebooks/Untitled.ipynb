```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 4.5, 0.005)
y = np.arange(-1, 4.5, 0.005)
X, Y = np.meshgrid(x, y)
Z = np.cos(2*np.pi*(np.sqrt(2)*Y)/2 )*np.cos(2*np.pi*(X-Y)/2)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z)

plt.show()
```
